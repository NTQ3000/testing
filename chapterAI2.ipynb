```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Dữ liệu
x = np.array([30, 40, 50, 60, 70, 80])
y = np.array([60, 75, 105, 115, 140, 150])

# Khởi tạo tham số
m = -1  # Hệ số góc
b = 15  # Hệ số chặn
n = 0.0001  # Tốc độ học

# Số lần lặp
iterations = 100

# Huấn luyện mô hình
for rep in range(iterations):
    dm = 0
    db = 0
    n_samples = len(x)
    
    for i in range(n_samples):
        # Tính độ dốc
        dm += -2/n_samples * (x[i] * (y[i] - (m * x[i] + b)))
        db += -2/n_samples * (y[i] - (m * x[i] + b))
    
    # Cập nhật tham số
    m -= n * dm
    b -= n * db
    
    # Hiển thị biểu đồ
    clear_output(wait=True)
    plt.scatter(x, y, color='blue', label='Dữ liệu gốc')
    plt.plot(x, m * x + b, color='red', label='Đường hồi quy')

    equation_text = f"h(x) = {m:.2f} * x + {b:.2f}"
    plt.text(0.5, 0.1, equation_text, transform=plt.gca().transAxes, fontsize=12,
             verticalalignment='bottom', horizontalalignment='center', color='black')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Hồi quy tuyến tính')
    plt.legend()
    plt.show()

# plt.scatter(X, y, color='blue', label='Dữ liệu gốc')
# plt.plot(X_test, y_pred, color='red', linewidth=2, label='Dự đoán')
# plt.xlabel('X')
# plt.ylabel('y')
# plt.title('Mô hình hồi quy tuyến tính')
# plt.legend()
# plt.show()

# In kết quả cuối cùng
print(f"Hệ số góc (m): {m}")
print(f"Hệ số chặn (b): {b}")
```
